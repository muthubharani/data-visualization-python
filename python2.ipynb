```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\Users\MUTHUBHARANI\Desktop\jupyterR\FoodData.csv'
df = pd.read_csv(file_path)
```

```python
df['Type'] = df['Type'].str.split(',')
type_df = df.explode('Type')
```

```python
type_df['Type'] = type_df['Type'].str.strip()
```

```python
type_counts = type_df['Type'].value_counts()
```

```python
plt.figure(figsize=(12, 8))
type_counts.sort_index().plot(kind='line', marker='o', color='green')
plt.title('Frequency of Types in Food Products')
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
